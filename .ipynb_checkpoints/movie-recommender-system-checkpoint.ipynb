```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on="title")
```

# Recommendation based on weighted average score

```python
R = movies["vote_average"]
v = movies["vote_count"]
m = movies["vote_count"].quantile(0.7)
C = movies["vote_average"].mean()
```

```python
w = (R*v) + (C*m)/(v+m)
movies["scores"] = w
```

```python
scored_df = movies.sort_values("scores", ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['scores'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Most popular movies', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

scored_df.to_csv("Scores_data.csv")
```

# Recommendation based on content

```python
#selecting only important tages
movies = movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
        else:
            break
    return L
movies["cast"] = movies["cast"].apply(convert3)
```

```python
def convertcrew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
movies["crew"] = movies["crew"].apply(convertcrew)
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[["movie_id","title","tags"]]
print(new_df.head())
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words ="english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
#select important tags
#extract import feature from those tags
#remove spaces
#concatinate all the tags
#remove the spaces and join them
#lowercase them
#do stemming using nltk.stem.porter PorterStemmer
#convert them into countvectorizer
#calculate cosine_similarity
#get similarity for the given movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    movies_list = sorted(list(enumerate(similarity[movie_index])), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
